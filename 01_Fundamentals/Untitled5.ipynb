```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import pandas as pd

# Load model and tokenizer
model_name = "gpt2"
model = AutoModelForCausalLM.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)

def generate_text(prompt, temperature, top_p, max_tokens):
    """Generate text using the model with specified parameters."""
    inputs = tokenizer(prompt, return_tensors="pt")
    outputs = model.generate(
        inputs.input_ids,
        do_sample=True,
        max_length=len(inputs.input_ids[0]) + max_tokens,
        temperature=temperature,
        top_p=top_p,
        pad_token_id=tokenizer.eos_token_id
    )
    return tokenizer.decode(outputs[0], skip_special_tokens=True)

# Define use cases and their parameters
use_cases = [
    {"name": "Code Generation", "prompt": "# Write a Python function to calculate factorial.", "temperature": 0.2, "top_p": 0.9, "max_tokens": 100},
    {"name": "Creative Writing", "prompt": "Write a poem about the beauty of nature.", "temperature": 0.7, "top_p": 0.95, "max_tokens": 150},
    {"name": "Chatbot Responses", "prompt": "User: What's the weather like today?\nChatbot:", "temperature": 0.5, "top_p": 0.9, "max_tokens": 100},
    {"name": "Code Comment Generation", "prompt": "# This function calculates the sum of an array\ndef calculate_sum(array):", "temperature": 0.3, "top_p": 0.8, "max_tokens": 200},
    {"name": "Story Telling", "prompt": "Once upon a time in a distant galaxy,", "temperature": 0.9, "top_p": 0.95, "max_tokens": 150},
    {"name": "Technical Explanation", "prompt": "Explain the concept of machine learning in simple terms.", "temperature": 0.4, "top_p": 0.85, "max_tokens": 300}
]

# Generate and record observations
results = []

for case in use_cases:
    generated_text = generate_text(case["prompt"], case["temperature"], case["top_p"], case["max_tokens"])
    results.append({
        "Use Case": case["name"],
        "Temperature": case["temperature"],
        "Top-P": case["top_p"],
        "Max Tokens": case["max_tokens"],
        "Observations": generated_text
    })

# Display results in a table
results_df = pd.DataFrame(results)
print(results_df.to_string(index=False))

# Optionally, save results to a CSV file
results_df.to_csv("use_case_results.csv", index=False)
```
