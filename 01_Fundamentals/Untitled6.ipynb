```python
n = int(input("Enter the number of values for the list: "))

# Create a list with n zeros
door = [0] * n
door
```

```python
for i in range (1, n+1): # monkey
  print(i)
  for j in range (1,n+1): #door
    if j%i==0:
      if door[j-1]==0:
        door[j-1]=1
      else:
        door[j-1]=0


door
```

```python
import math
def bulbSwitch(n: int) -> int:
    # The k-th bulb can only be switched when k % i == 0.
    # So, we can rephrase the problem:
    # To find number of numbers <= n that have odd factors.
    # Obviously, only square numbers have odd factor(s).
    # e.g. n = 10, only 1, 4, and 9 are square numbers that <= 10
    return math.isqrt(n)
```
