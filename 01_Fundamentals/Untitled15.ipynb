```python
# Install required libraries
!pip install --upgrade transformers accelerate bitsandbytes

# Import libraries
from transformers import AutoTokenizer, AutoModelForCausalLM
from torch.cuda import empty_cache

# Load the model and tokenizer with 8-bit precision
model_name = "Salesforce/codegen-2B-mono"  # Lightweight LLM for coding tasks
tokenizer = AutoTokenizer.from_pretrained(model_name)

model = AutoModelForCausalLM.from_pretrained(
    model_name,
    load_in_8bit=True,   # Memory-efficient loading
    device_map="auto"    # Automatically map to GPU
)

# Function to generate code or responses
def generate_output(prompt, max_tokens=100, temperature=0.7):
    """
    Generate output using the loaded model.

    Parameters:
    - prompt (str): The input prompt to guide the generation.
    - max_tokens (int): Maximum number of tokens to generate.
    - temperature (float): Controls randomness in generation.

    Returns:
    - str: Generated output.
    """
    input_ids = tokenizer(prompt, return_tensors="pt").input_ids.to("cuda")
    output = model.generate(
        input_ids,
        max_new_tokens=max_tokens,
        temperature=temperature,
        pad_token_id=tokenizer.eos_token_id
    )
    return tokenizer.decode(output[0], skip_special_tokens=True)

# Code Generation
def generate_code(task_description):
    """
    Generate Python code based on the task description.

    Parameters:
    - task_description (str): The description of the coding task.

    Returns:
    - str: Generated Python code.
    """
    prompt = f"Write Python code for the following task:\n{task_description}"
    return generate_output(prompt, max_tokens=150)

# Debugging
def debug_code(code_snippet):
    """
    Debug Python code and provide the corrected version.

    Parameters:
    - code_snippet (str): The Python code to debug.

    Returns:
    - str: Debugged Python code.
    """
    prompt = f"Debug the following Python code and fix any errors:\n{code_snippet}"
    return generate_output(prompt, max_tokens=200)

# Test Case Generation
def generate_test_cases(function_description):
    """
    Generate test cases for a given function.

    Parameters:
    - function_description (str): Description of the function to test.

    Returns:
    - str: Python test cases for the function.
    """
    prompt = f"Write Python test cases for the following function:\n{function_description}"
    return generate_output(prompt, max_tokens=150)

# Example Usage
if __name__ == "__main__":
    # Task 1: Code Generation
    task_description = "Create a function to calculate the factorial of a number."
    print("### Code Generation ###")
    generated_code = generate_code(task_description)
    print(generated_code)

    # Task 2: Debugging
    buggy_code = """
def calculate_factorial(n):
    if n == 0:
        return 1
    else:
        return n * calculate_factorial(n - 1)
"""
    print("\n### Debugging ###")
    debugged_code = debug_code(buggy_code)
    print(debugged_code)

    # Task 3: Test Case Generation
    function_description = "A function that calculates the factorial of a number."
    print("\n### Test Case Generation ###")
    test_cases = generate_test_cases(function_description)
    print(test_cases)

    # Clean up GPU memory
    empty_cache()
```
