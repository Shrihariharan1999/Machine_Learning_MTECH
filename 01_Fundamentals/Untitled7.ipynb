```python
!pip install faiss-cpu
from sentence_transformers import SentenceTransformer
import faiss
import numpy as np

# Initialize model and vector database
model = SentenceTransformer('all-MiniLM-L6-v2')
d = 384  # Embedding dimension
index = faiss.IndexFlatL2(d)

# Example data
documents = [
    "This is a document about machine learning.",
    "An introduction to neural networks.",
    "Exploring deep learning for AI."
]
doc_embeddings = model.encode(documents)
index.add(np.array(doc_embeddings, dtype='float32'))

# Query
query = "Tell me about deep learning"
query_embedding = model.encode([query])
D, I = index.search(np.array(query_embedding, dtype='float32'), k=2)

# Results
for idx in I[0]:
    print(f"Matched Document: {documents[idx]}")
```
