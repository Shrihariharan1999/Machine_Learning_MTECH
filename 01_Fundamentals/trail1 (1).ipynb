```python
numerical=['person_age','person_gender','person_education','person_income','person_emp_exp','loan_amnt','loan_int_rate','loan_percent_income','cb_person_cred_hist_length',
           'credit_score']
cat=['person_home_ownership','loan_intent','previous_loan_defaults_on_file']
```

```python
numeric_transformer=StandardScaler()
categorical_transformer= OrdinalEncoder()
```

```python
from sklearn.compose import ColumnTransformer
```

```python
from sklearn.pipeline import Pipeline
```

```python
preprocessor=ColumnTransformer(transformers=[ ('numeric',numeric_transformer,numerical),('cate',categorical_transformer,cat)])
```

```python
pipeline=Pipeline([('preprocessor',preprocessor),('classifier',RandomForestClassifier())])
```

```python
X_train.shape, y_train.shape
```

```python
pipeline.fit(X_train,y_train)
```

```python
y_pred =pipeline.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_pred,y_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm=confusion_matrix(y_pred,y_test)
```

```python
# Plotting confusion matrix as a heatmap for better visualization
sns.heatmap(cm, annot=True,fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
