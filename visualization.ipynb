```python
plt.Figure(figsize=(10,20))
plt.hist(df['Age'], color='blue', ec='black', bins=15)
plt.xlabel("Age")
plt.ylabel("count")
plt.title( "Age Distribution")
```

```python
sns.displot(df['BMI'], kde=True, bins=15)
```

```python
a=df['Target'].value_counts()
a
```

```python
plt.Figure(figsize=(20,20))
plt.pie(a,labels=a.index,autopct="%1.1f%%")
```
