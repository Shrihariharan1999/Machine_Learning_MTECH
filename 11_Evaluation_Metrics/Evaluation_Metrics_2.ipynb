```python
import nltk
from nltk.translate.bleu_score import corpus_bleu, SmoothingFunction
from rouge_score import rouge_scorer
from bert_score import score

# Download NLTK resources
nltk.download('wordnet')
nltk.download('punkt')

# Example data
references = [
    ["The sun sets in the west."], # Reference for prediction 1
    ["A gentle breeze flows through the trees."] # Reference for prediction 2
]
predictions = [
    "The sun goes down in the west.",  # Prediction 1
    "A soft breeze moves through the trees."  # Prediction 2
]

# BLEU score
smooth = SmoothingFunction().method1
bleu_score = corpus_bleu(
    [[ref[0].split()] for ref in references],  # Tokenize references
    [pred.split() for pred in predictions],    # Tokenize predictions
    smoothing_function=smooth
)

# ROUGE score
scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
rouge_scores = [scorer.score(ref[0], pred) for ref, pred in zip(references, predictions)]

# BERTScore
P, R, F1 = score(predictions, [ref[0] for ref in references], lang="en")

# Print results
print(f"BLEU Score: {bleu_score}")
print("ROUGE Scores:")
for i, scores in enumerate(rouge_scores):
    print(f"Sample {i + 1}: {scores}")
print(f"BERTScore (F1): {F1.mean().item():.4f}")
```
