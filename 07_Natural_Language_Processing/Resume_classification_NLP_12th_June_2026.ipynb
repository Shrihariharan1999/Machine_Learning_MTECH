````python
X=df["Resume"]
y=df["Category"]
````

````python
from sklearn.preprocessing import LabelEncoder
from sklearn.svm import LinearSVC
from sklearn.feature_extraction.text import TfidfVectorizer #TF-IDF
````

````python
from sklearn.model_selection import train_test_split
from sklearn.metrics import (accuracy_score,classification_report,confusion_matrix)
````

````python
def preprocess(text):
  text=text.lower()
  text=re.sub(r'[^A-Za-z0-9 ]',"",text)
  text = re.sub(r'\s+', ' ', text)
  return text.strip()
````

````python
X
````

````python
X=X.apply(preprocess)
````

````python
X
````

````python
le=LabelEncoder()
y=le.fit_transform(y)
````

````python
le.classes_
````

````python
y
````

````python
inverse_mapping = dict(zip(range(len(le.classes_)), le.classes_))
print(inverse_mapping)
````

````python
tfidf = TfidfVectorizer(
    max_features=5000,
    ngram_range=(1,2)
)

X_vec= tfidf.fit_transform(X)
````

````python
X_vec.shape
````

````python
X.shape
````

````python
X_train, X_test, y_train, y_test = train_test_split(
    X_vec,
    y,
    test_size=0.2,
    random_state=42,
    stratify=y
)
````

````python
model = LinearSVC()

model.fit(X_train, y_train)
````

````python
pred_linear_svc= model.predict(X_test)
````

````python
print("Accuracy")

print(
    accuracy_score(y_test, pred_linear_svc)
)
````

````python
print(
    classification_report(
        y_test,
        pred_linear_svc,
        target_names=le.classes_
    )
)
````

````python
cm = confusion_matrix(
    y_test,
    pred_linear_svc
)

print(cm)
````

````python
sentences = X_tokenize.tolist()
````

````python
w2v_model = Word2Vec(
    sentences=sentences,
    vector_size=100,
    window=5,
    min_count=2,
    workers=4,
    sg=1
)
````

````python
def document_vector(words):

    word_vectors = []

    for word in words:

        if word in w2v_model.wv:
            word_vectors.append(w2v_model.wv[word])

    if len(word_vectors) == 0:

        return np.zeros(w2v_model.vector_size)

    return np.mean(word_vectors,axis=0)
````

````python
X_vec = np.array([document_vector(tokens) for tokens in X_tokenize])
````

# Resume Classification - NLP Model Comparison

## Dataset

- Multi-class Resume Classification Dataset
- Classes:
  - Python Developer
  - Java Developer
  - Web Developer
  - Project Manager
  - Database Administrator
  - Security Analyst
  - Network Administrator
  - Systems Administrator
  - Software Developer
  - Front-End Developer

---

# Models Evaluated

| Model | Accuracy |
|---------|---------:|
| Word2Vec + GaussianNB | 71.8% |
| Word2Vec + Logistic Regression | 82.4% |
| FastText + Logistic Regression | 82.9% |
| TF-IDF + LinearSVC | 93.4% |

---

# Key Findings

## 1. TF-IDF + LinearSVC performed best

Accuracy:

```text
93.4%
```

Reasons:

- Resume classification is heavily keyword-driven.
- Important technologies such as:
  - Python
  - TensorFlow
  - AWS
  - Docker
  - Spring
  - Kubernetes

  are preserved as individual features.

- LinearSVC works exceptionally well on:
  - High-dimensional
  - Sparse
  - TF-IDF feature spaces

---

## 2. Word2Vec improved significantly with Logistic Regression

Accuracy:

```text
71.8% → 82.4%
```

Comparison:

| Model | Accuracy |
|---------|---------:|
| Word2Vec + GaussianNB | 71.8% |
| Word2Vec + Logistic Regression | 82.4% |

Reason:

- GaussianNB assumes feature independence and Gaussian distributions.
- Word embeddings contain correlated continuous features.
- Logistic Regression handles embedding vectors more effectively.

---

## 3. FastText slightly outperformed Word2Vec

Accuracy:

```text
Word2Vec : 82.4%
FastText : 82.9%
```

Improvement:

```text
≈ 0.5%
```

Reason:

- FastText uses character n-grams.
- Can generate embeddings for rare or unseen words.
- Better handling of:
  - Python2
  - Python27
  - Python3x
  - Misspellings
  - Rare skills

---

## 4. FastText Learned Subword Relationships

Most similar words to:

```text
python
```

included:

```text
python2
python27
python3x
modpython
rubypython
javapython
```

Inference:

- FastText groups words using character structure.
- Demonstrates successful subword learning.

---

## 5. Word2Vec Learned Context-Based Relationships

Most similar words to:

```text
python
```

included:

```text
perl
lua
erlang
```

Inference:

- Word2Vec groups words appearing in similar contexts.
- Learns semantic similarity rather than character similarity.

---

---

# Confusion Matrix Insights

Most common misclassifications:

```text
Software Developer → Web Developer
Network Administrator → Systems Administrator
Front-End Developer → Web Developer
```

Reason:

- Similar technologies and skills appear across these roles.
- Resume content overlaps considerably.

---

# Overall Conclusion

TF-IDF + LinearSVC remains the strongest model for this resume classification dataset because it preserves important keywords explicitly.

Word2Vec and FastText successfully learn semantic and subword relationships but lose some discriminative information when document embeddings are created using simple averaging.

The experiment demonstrates that:

- Better embeddings do not always lead to better classification accuracy.
- For keyword-heavy datasets, TF-IDF remains a highly competitive baseline.
- FastText provides better handling of rare and unseen words compared to Word2Vec.
- Model selection should depend on the nature of the dataset rather than assuming newer embeddings will always perform better.
