```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import nltk
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer

# Download required resources from nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Initialize stemming and lemmatization
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()
```

```python
train=pd.read_csv("/content/train.csv")
test=pd.read_csv("/content/test.csv")
df=pd.concat([train,test])
df.head()
```

```python
df['edited_text'] = df['Tweet'].apply(lambda x: ' '.join([stemmer.stem(word) for word in word_tokenize(x)]))
```

```python
df
```

```python
df['edited_text'] = df['edited_text'].apply(lambda x: ' '.join([lemmatizer.lemmatize(word) for word in word_tokenize(x)]))
```

```python
df
```

```python
df['edited_text'] = df['edited_text'].apply(lambda x: ' '.join([word for word in x.split() if word not in stopwords.words('english')]))
```

```python
df
```

```python
df['edited_text'] = df['edited_text'].apply(lambda x: re.sub(r'[^\w\s]', '', x))
```

```python
df
```

```python
from nltk.tokenize import word_tokenize
```

```python
df['edited_text'] = df['edited_text'].apply(word_tokenize)
```

```python
df['Category'].replace('Tweet', 'neutral', inplace=True)
```

```python
category_mapping = {
    'negative': -1,
    'neutral': 0,
    'positive': 1
}
```

```python
df['Category_Encoded'] = df['Category'].map(category_mapping)
```

```python
X=train['edited_text']
y=train['Category_Encoded']
```

```python
# Convert each element of X_train to a string
import pandas as pd
from sklearn.model_selection import train_test_split
X = X.astype(str)
X_train, X_test_S, y_train, y_test_S = train_test_split(X,y,test_size=0.3,random_state = 0)
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.linear_model import SGDClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

pipeline_sgd = Pipeline([
    ('vect', CountVectorizer()),
    ('tfidf',  TfidfTransformer()),
    ('clf', SGDClassifier(random_state=0)),
])

model = pipeline_sgd.fit(X_train, y_train)
```

```python
X_train
```

```python
from sklearn.metrics import classification_report
y_predict_S = model.predict(X_test_S)
print(classification_report(y_test_S, y_predict_S))
```

```python
from sklearn.metrics import accuracy_score # import the module

accuracy =accuracy_score(y_test_S, y_predict_S)
print("Accuracy:", accuracy)
```
