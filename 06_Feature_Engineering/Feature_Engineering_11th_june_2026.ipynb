```python
from scipy.stats import ttest_ind
import pandas as pd

results = []

for col in num_cols:

    group1 = df[df["Attrition"] == 0][col]
    group2 = df[df["Attrition"] == 1][col]

    mean_stayed = group1.mean()
    mean_left = group2.mean()

    stat, p = ttest_ind(
        group1,
        group2,
        nan_policy="omit"
    )

    results.append([
        col,
        round(mean_stayed, 2),
        round(mean_left, 2),
        round(stat, 2),
        round(p, 6)
    ])

ttest_df = pd.DataFrame(
    results,
    columns=[
        "Feature",
        "Mean_Stayed",
        "Mean_Left",
        "T_Statistic",
        "P_Value"
    ]
).sort_values("P_Value")

print(ttest_df)
```

# T-Test Insights: Employee Attrition Analysis

## Key Findings

1. **Employees who left the company are younger on average** (33.61 years) compared to employees who stayed (37.56 years), indicating younger employees have a higher attrition tendency.

2. **Monthly Income is a strong attrition indicator.** Employees who left earned significantly less (4787) than employees who stayed (6833), suggesting compensation plays an important role in retention.

3. **Less experienced employees are more likely to leave.** Employees who left had fewer total working years (8.24) compared to employees who stayed (11.86).

4. **Lower job-level employees show higher attrition.** The average Job Level for employees who left was 1.64 compared to 2.15 for employees who stayed.

5. **Employees with shorter tenure are more likely to leave.** Employees who left had spent fewer years at the company (5.13 years) compared to employees who stayed (7.37 years).

6. **Employees who left had fewer years in their current role and with their current manager**, indicating that stronger role familiarity and manager relationships are associated with retention.

7. **Stock options appear to improve employee retention.** Employees who stayed had a higher average Stock Option Level (0.85) than employees who left (0.53).

8. **Job Satisfaction and Environment Satisfaction are significantly lower among employees who left**, suggesting employee engagement and workplace conditions influence attrition.

9. **Employees living farther from the workplace tend to leave more often.** The average Distance From Home for employees who left was 10.63 compared to 8.92 for employees who stayed.

10. **Performance Rating, Hourly Rate, Monthly Rate, and Percent Salary Hike do not show a statistically significant relationship with attrition**, indicating these factors alone are poor predictors of employee turnover.

## Overall Conclusion

Employees most likely to leave are generally younger, lower-paid, less experienced, lower in job level, newer to the organization, less satisfied with their jobs and work environment, and located farther from the workplace. Compensation, experience, tenure, satisfaction, and organizational engagement appear to be the strongest drivers of employee retention in this dataset.


```python
#Chi-Square
from scipy.stats import chi2_contingency
import pandas as pd

results = []

for col in cat_cols:

    table = pd.crosstab(df[col], df["Attrition"])

    chi2, p, dof, exp = chi2_contingency(table)

    results.append([
        col,
        round(chi2, 2),
        dof,
        round(p, 6),
        "Significant" if p < 0.05 else "Not Significant"
    ])

chi_df = pd.DataFrame(
    results,
    columns=[
        "Feature",
        "Chi_Square",
        "DOF",
        "P_Value",
        "Decision"
    ]
).sort_values("P_Value")

print(chi_df)
```

# Chi-Square & Crosstab Analysis: Employee Attrition

## 1. OverTime (χ² = 87.56, p < 0.001)

| Category | Attrition Rate |
| -------- | -------------- |
| No       | 10.44%         |
| Yes      | 30.53%         |

### Interpretation

* Employees working overtime have nearly **3 times higher attrition** than employees not working overtime.
* This feature has the **highest statistical association** with attrition among the categorical variables analyzed.
* The extremely small p-value confirms that this difference is highly unlikely to have occurred by chance.

---

## 2. JobRole (χ² = 86.19, p < 0.001)

### Highest Attrition Roles

* Sales Representative → **39.76%**
* Laboratory Technician → **23.94%**
* Human Resources → **23.08%**

### Lowest Attrition Roles

* Research Director → **2.50%**
* Manager → **4.90%**
* Healthcare Representative → **6.87%**

### Interpretation

* Attrition varies substantially across job roles.
* Sales Representatives experience the highest turnover rate, while Research Directors and Managers exhibit the strongest retention.
* The large Chi-Square statistic confirms that job role is strongly associated with attrition.

---

## 3. MaritalStatus (χ² = 46.16, p < 0.001)

| Category | Attrition Rate |
| -------- | -------------- |
| Single   | 25.53%         |
| Married  | 12.48%         |
| Divorced | 10.09%         |

### Interpretation

* Single employees have approximately **double the attrition rate** of married employees.
* Married and divorced employees tend to remain with the company longer.
* The Chi-Square test confirms that marital status has a significant relationship with attrition.

---

## 4. BusinessTravel (χ² = 24.18, p < 0.001)

| Category          | Attrition Rate |
| ----------------- | -------------- |
| Travel_Frequently | 24.91%         |
| Travel_Rarely     | 14.96%         |
| Non-Travel        | 8.00%          |

### Interpretation

* Employees who travel frequently have the highest attrition rate.
* Non-travel employees show the lowest turnover.
* Frequent travel appears to increase the likelihood of employee departure.
* The Chi-Square test confirms a statistically significant relationship.

---

## 5. EducationField (χ² = 16.02, p = 0.0068)

### Highest Attrition Fields

* Human Resources → **25.93%**
* Technical Degree → **24.24%**
* Marketing → **22.01%**

### Lowest Attrition Fields

* Other → **13.41%**
* Medical → **13.58%**
* Life Sciences → **14.69%**

### Interpretation

* Attrition differs across educational backgrounds.
* Employees from Human Resources and Technical Degree backgrounds show higher turnover rates.
* The Chi-Square test indicates a statistically significant association.

---

## 6. Department (χ² = 10.80, p = 0.0045)

| Department             | Attrition Rate |
| ---------------------- | -------------- |
| Sales                  | 20.63%         |
| Human Resources        | 19.05%         |
| Research & Development | 13.84%         |

### Interpretation

* Sales department employees exhibit the highest attrition.
* Research & Development employees show the lowest attrition.
* Department is statistically associated with employee turnover.

---

## 7. Gender (χ² = 1.12, p = 0.2906)

| Gender | Attrition Rate |
| ------ | -------------- |
| Male   | 17.01%         |
| Female | 14.80%         |

### Interpretation

* Although male employees show slightly higher attrition, the difference is small.
* The p-value is greater than 0.05, indicating that the observed difference can reasonably occur due to random variation.
* Gender does not have a statistically significant relationship with attrition.

---

# Overall Conclusions

1. OverTime is the strongest categorical predictor of attrition.
2. Sales Representatives experience the highest employee turnover among all job roles.
3. Single employees are significantly more likely to leave than married or divorced employees.
4. Frequent business travel is associated with higher attrition.

```python
#Mutual Information
from sklearn.feature_selection import mutual_info_classif
```

```python
X = df.drop("Attrition", axis=1)
y = df["Attrition"]
```

```python
X = pd.get_dummies(X, drop_first=True)
```

```python
mi = mutual_info_classif(X, y, random_state=42)

mi_df = pd.DataFrame({
    "Feature": X.columns,
    "MI_Score": mi
}).sort_values("MI_Score", ascending=False)

print(mi_df.head(20))
```

## Mutual Information Analysis

- JobLevel (0.0344) is the most informative feature for predicting employee attrition.
- OverTime_Yes (0.0343) is one of the strongest indicators of employee turnover.
- MaritalStatus_Single (0.0336) suggests that single employees have a higher attrition tendency.
- MonthlyIncome (0.0320) is strongly associated with employee retention.
- YearsInCurrentRole (0.0310) indicates that employees with shorter tenure in their current role are more likely to leave.
- TotalWorkingYears (0.0249) and Age (0.0208) provide meaningful predictive information about attrition.
- Mutual Information confirms the findings from T-Test and Chi-Square analyses, highlighting JobLevel, OverTime, MonthlyIncome, and MaritalStatus as key attrition drivers.
- Features with higher MI scores contribute more information toward predicting employee attrition and are strong candidates for model building.

```python
#RFE (Recursive Feature Elimination)
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(max_iter=1000)
```

```python
rfe = RFE(
    estimator=model,
    n_features_to_select=10
)
```

```python
rfe.fit(X, y)
```

```python
rfe_df = pd.DataFrame({
    "Feature": X.columns,
    "Selected": rfe.support_,
    "Rank": rfe.ranking_
})

rfe_df.sort_values("Rank")
```

## RFE Analysis

- RFE selected features that collectively provide the best Logistic Regression performance.
- OverTime, MaritalStatus, BusinessTravel, and several JobRole variables were retained, indicating strong predictive value.
- Features such as MonthlyIncome, Age, and TotalWorkingYears were removed despite being important in previous analyses because they share information with other selected features.
- RFE removes redundant predictors and focuses on the optimal feature subset rather than individual feature strength.
- Different feature selection methods may produce different rankings because each method evaluates importance using different criteria.
- Features consistently identified across multiple methods (OverTime, MaritalStatus, BusinessTravel) can be considered highly reliable predictors of attrition.
