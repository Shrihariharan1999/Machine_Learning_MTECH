```python
import pandas as pd
import numpy as np
```

```python
tr=pd.read_csv('/content/train.csv')
```

```python
te=pd.read_csv("/content/test.csv")
```

```python
frames = [tr,te]
df = pd.concat(frames)
```

```python
df.head()
```

```python
df.info(),tr.info(),te.info()
```

```python
df.info()
```

```python
df["Drug"].fillna(df["Drug"].mode()[0],inplace=True)
df.drop("Ascites",axis=1,inplace=True)
df.drop("Hepatomegaly",axis=1,inplace=True)
df.drop("Spiders",axis=1,inplace=True)
df["Bilirubin"].fillna(df["Bilirubin"].mean(),inplace=True)
df["Cholesterol"].fillna(df["Cholesterol"].mean(),inplace=True)
df.drop("Copper",axis=1,inplace=True)
df.drop("Alk_Phos",axis=1,inplace=True)
df.drop("SGOT",axis=1,inplace=True)
df.drop("Tryglicerides",axis=1,inplace=True)
df["Platelets"].fillna(df["Platelets"].mean(),inplace=True)
df["Prothrombin"].fillna(df["Prothrombin"].mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df
```

```python
tr.shape, te.shape
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df["Drug"]=le.fit_transform(df["Drug"])
df["Sex"]=le.fit_transform(df["Sex"])
df["Edema"]=le.fit_transform(df["Edema"])
df["Bilirubin"]=le.fit_transform(df["Bilirubin"])
df["Status"].replace({"C":0,"CL":1,"D":2},inplace=True)
```

```python
df
```

```python
train=df.iloc[:15000]
test=df.iloc[15000:]
```

```python
train
```

```python
test
```

```python
train.shape,tr.shape,test.shape,te.shape
```

```python
train=train.drop("id",axis=1)
test=test.drop("id",axis=1)
```

```python
x=train.drop("Status",axis=1)
y=train["Status"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
(x_train,x_test,y_train,y_test)=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log = LogisticRegression(multi_class='multinomial')
log.fit(x_train, y_train)
```

```python
y_pred_sample = log.predict(x_test)
y_pred_sample
```

```python
pd.DataFrame(y_pred_sample)
```

```python
y_pred_proba_sample = log.predict_proba(x_test)
```

```python
pd.DataFrame(y_pred_proba_sample)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,median_absolute_error,accuracy_score
accuracy_score(y_pred_sample, y_test)
```

```python
from sklearn.metrics import log_loss
log_loss(y_test,y_pred_proba_sample)
```
