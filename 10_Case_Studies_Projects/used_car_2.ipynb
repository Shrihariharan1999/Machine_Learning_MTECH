```python
x_sample=train.drop("Price",axis=1)
x_sample=x_sample.drop("Id",axis=1)
y_sample=train["Price"]
```

```python
x_sample
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
x_train_sample,x_test_sample,y_train_sample,y_test_sample = train_test_split(x_sample,y_sample,test_size=0.2,random_state=41)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression(copy_X=True, fit_intercept=True, n_jobs=None)
lr.fit(x_train_sample,y_train_sample)
```

```python
y_pred_sample =lr.predict(x_test_sample)
print(y_pred_sample)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test_sample,y_pred_sample)
```
